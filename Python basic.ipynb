6. Why are lists mutable, but tuples are immutable?

 - When list is modified a mutable object, the object itself is altered in place, and its memory address (identity) remains the same.
 - A tuples immutability means that once it is created and its elements are set, it cannot be changed or altered in place in memory.

10.  What is the difference between implicit and explicit type casting?
- 	The difference between implicit and explicit type casting is:
- Implicit Type Casting:
     - Implicit Type Casting is done automatically by python it converts one data type to another without you needing to do anything.
     - It  happens when Python needs to perform an operation involving two different data types. Example: when you add an integer and a float, python converts the integer to float automatically.
     
- Explicit Type Casting:
     -  Explicit Type Casting needs to be converted manually by user, the user needs to convert one data type to another manually using built-in functions like int, float, str, etc.
     - It needs to be done when we want specific type conversion. Example: Converting a string "123" to an integer 123 manually.

```python
# 14. Write a program to reverse a string using a while loop.
# answer:

text = "Vivek"
reverse = ""
i = len(text) - 1

while i >= 0:
    reverse += text[i]
    i -= 1

print("Reversed string:", reverse)
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
# answer:

num = int(input("Enter a number: "))
if num < 0:
    print("Factorial is not defined for negative numbers.")
else:
    factorial = 1
    i = 1
    while i <= num:
        factorial *= i
        i += 1
    print("Factorial of", num, "is:", factorial)
```
