6.  Explain the concept of generators in Python and how they are defined.
- Generators are a simple way to create iterators using functions and the yield keyword instead of returning values.
- Defined using the 'def' keyword.
- Use 'yeild' instead of 'return'.
- Automatically implement the iterator protocol __iter__ and __next__.
- Do not store all values in memory; values are generated on-the-fly
- Lazy Evaluation: Generates values on demand instead of computing all at once.
- Memory Efficient: Does not store entire sequences in memory.
- Iterator Protocol: A generator automatically implements __iter__() and __next__() methods.


```python
def my_generator():
    yield 1
    yield 2
    yield 3

#Calling the generator:
gen = my_generator()
print(next(gen))
print(next(gen))
```

7. What are the advantages of using generators over regular functions?
 - Memory Efficiency
   - Generators do not store the entire sequence in memory. Instead, they generate values one at a time using yield, which is ideal for handling large datasets or infinite sequences.
 - Lazy Evaluation
   - Generators compute values only when requested. This "on-demand" approach improves performance and reduces unnecessary computation.
 - Improved Performance
   - Since generators avoid loading entire data structures into memory, they often execute faster in scenarios involving large loops or data streams.
 - Simplified Code for Iteration
   - Generators provide a clean and readable way to write custom iterators without manually implementing __iter__() and __next__() methods.
 - State Preservation
   - Each time a generator is paused with yield, it retains its internal state. When resumed, it continues from where it left off, making it suitable for tasks like parsing or streaming.
 - Support for Infinite Sequences
   - Generators can model infinite sequences (e.g., Fibonacci numbers, data streams) without causing memory overflow.

9.  Explain the purpose and usage of the `map()` function in Python
- The map function applies a given function to all items in an input iterable (like a list) and returns an iterator with the results.
- Usage:
  - The map() function in Python is used to apply a given function to each item of an iterable (like a list, tuple, etc.) and return a map object (iterator) containing the results.
  - It is a functional programming tool that helps eliminate the need for explicit for loops.
- Transformation: Used to transform each item in an iterable by applying the specified function.
- Syntax: map(function, iterable)
  - function : A function to apply to each element.
  - iterable : A sequence (e.g., list, tuple) whose elements will be passed to the function.


```python
def square(x):
    return x * x

numbers = [1, 2, 3, 4]
result = map(square, numbers)
print(list(result))
```

- Advantages
  - Improves code readability and conciseness.
  - Avoids explicit loops for element-wise operations.
  - Efficient for large datasets due to lazy evaluation.


10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?
- map()
   - Purpose: Transforms each element of an iterable using a given function.
   - Returns: A map object (iterator) of transformed elements.
   - Usage: When you want to apply a function to every item.


```python
nums = [1, 2, 3, 4]
result = map(lambda x: x*2, nums)
print(list(result))
```

- filter()
  - Purpose: Selects elements from an iterable based on a condition (Boolean test).
  - Returns: A filter object (iterator) containing only elements where the function returned True.
  - Usage: When you want to keep only certain items.


```python
nums = [1, 2, 3, 4, 5, 6]
result = filter(lambda x: x % 2 == 0, nums)
print(list(result))
```

- reduce()
  - Purpose: Repeatedly applies a function to the elements of an iterable to reduce it to a single cumulative value.
  - Returns: A single value.
  - Usage: When you want to combine all items into one (sum, product, etc.).
Note: reduce() is part of the functools module.


```python
from functools import reduce

nums = [1, 2, 3, 4]
result = reduce(lambda x, y: x * y, nums)
print(result)
```

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

```python
#solution:
def sum_of_evens(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total


nums = [10, 21, 32, 43, 54]
print("Sum of even numbers:", sum_of_evens(nums))
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
#solution:
def is_prime(n):
    if n < 2 or n > 200:
        return False  # Outside valid range
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

for num in range(1, 201):
    if is_prime(num):
        print(f"{num} is prime")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

```python
#solution:
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.a, self.b = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        value = self.a
        self.a, self.b = self.b, self.a + self.b
        self.count += 1
        return value

fib = FibonacciIterator(10)
for num in fib:
    print(num, end=' ')
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
#solution:
def powers_of_two(limit):
    for exp in range(limit + 1):
        yield 2 ** exp

for value in powers_of_two(5):
    print(value, end=" ")
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

```python
#solution:

# Function convert Celsius to Fahrenheit
def c_to_f(celsius):
    return (celsius * 9/5) + 32

# Temperatures list in Celsius
temps_c = [0, 20, 37, 100]


temps_f = list(map(c_to_f, temps_c))

print("Celsius:", temps_c)
print("Fahrenheit:", temps_f)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
#solution:
def remove_vowels(text):
    vowels = "aeiouAEIOU"

    return "".join(filter(lambda ch: ch not in vowels, text))


s = "Hello Vivek Thakare"
print("Original:", s)
print("Without vowels:", remove_vowels(s))
```
