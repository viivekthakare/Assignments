1. What is NumPy, and why is it widely used in Python
- NumPy (short for Numerical Python) is a foundational open-source library in Python designed for numerical and scientific computing. It was developed in 2005 by Travis Oliphant through the integration of two earlier libraries—Numeric and Numarray.
- At its core, NumPy introduces the ndarray (N-dimensional array), a powerful data structure that allows efficient storage and manipulation of homogeneous numerical data. These arrays can represent vectors, matrices, and higher-dimensional tensors.
- NumPy also provides a suite of mathematical functions, including:
  - Linear algebra operations
  - Fourier transforms
  - Random number generation
  - Statistical computations-
- Its internal implementation is written in optimized C code, which allows it to perform computations much faster than native Python structures like lists
- Reasons for Wide Usage

  - Performance: Operations are implemented in optimized C code, making them significantly faster than equivalent pure-Python loops.

  - Memory Efficiency: Arrays are stored in contiguous memory blocks, reducing overhead and improving cache performance.

  - Foundation for the Scientific Python Ecosystem: Many libraries in data science, machine learning, and scientific computing depend on NumPy as a core building block.

  - Ease of Use: Its syntax is concise and expressive, making it accessible to both beginners and experts.


2.  How does broadcasting work in NumPy
- Broadcasting in NumPy is a set of rules that allows arithmetic operations between arrays of different shapes without the need for explicit copying or looping. When two arrays of unequal dimensions are combined in an operation, NumPy “broadcasts” the smaller array across the larger one so that they behave as if they had compatible shapes.
- Broadcasting Rules:
  -	Alignment from the Trailing Dimensions: NumPy compares the shapes of the two arrays starting from the last (rightmost) dimension and moves leftward.
  -	Dimension Compatibility
       -	Two dimensions are compatible if:
       -	They are equal, or
       -	One of them is 1.
  -	If these conditions are not met for every dimension, broadcasting is not possible and an error is raised.
  -	Automatic Expansion: Whenever a dimension of size 1 is encountered, NumPy virtually replicates (without actual copying) that dimension to match the corresponding size of the other array.

4.  Explain the use of the groupby() method in Pandas
-	The groupby() method in Pandas is used to split a DataFrame into groups based on the values of one or more columns and then apply operations—such as aggregation, transformation, or filtering—on each group separately. It follows the Split–Apply–Combine strategy:
-	Split: Divide the DataFrame into groups according to specified column(s) or a function.
-	Apply: Perform an operation on each group, such as computing the mean, sum, count, or applying a custom function.
-	Combine: Merge the results of these operations back into a new DataFrame or Series.
-	Key Uses
  -	Aggregation: Summarizing data (e.g., .sum(), .mean(), .count()).
  -	Transformation: Returning an object of the same size where each group is transformed (e.g., normalizing within each group).
  -	Filtering: Selecting groups that meet certain conditions.

6. What are the differences between NumPy arrays and Python lists
- Data Type Handling
   - Python Lists:
	 - Can hold elements of mixed data types (e.g., integers, strings, floats together).
	 - Each element is a separate Python object.
   - NumPy Arrays:
   	- Store elements of a single, fixed data type (all integers, all floats, etc.).
   	- Enables efficient memory use and fast numerical operations.
- Performance
   - Python Lists:
     - Slower for numerical computations because operations are performed element by element using Python’s dynamic typing.
   - NumPy Arrays:
     - Significantly faster for mathematical calculations because they use contiguous memory blocks and vectorized operations implemented in C.
- Functionality
   - Python Lists:
     - General-purpose container with basic operations like append, remove, and slicing.
   - NumPy Arrays:
     - Provide a wide range of scientific and mathematical functions: broadcasting, linear algebra, random number generation, Fourier transforms, etc.
-	Memory Usage
  - Python Lists:
    - Store references to objects, leading to higher memory overhead.
  - NumPy Arrays:
    - Store actual data in a compact form, making them memory-efficient.
- Operations
   - Python Lists:
      - Arithmetic operations (e.g., list1 + list2) result in concatenation, not element-wise computation.
   - NumPy Arrays:
      - Support element-wise operations (array1 + array2 adds corresponding elements directly).

17. Explain the difference between apply() and map() in Pandas
- map()

   - Used With: Series only.

   - Purpose: Applies a function, dictionary, or mapping to each element of the Series.

   - Output: Returns a new Series of the same size

```python
import pandas as pd

s = pd.Series([1, 2, 3, 4])
squared = s.map(lambda x: x**2)
print(squared)
```

- apply()

   - Used With: Series or DataFrame.

   - Purpose: Applies a function along rows or columns (for DataFrame) or elements (for Series).

   - Output: Can return a Series, DataFrame, or scalar depending on the function.

   - Example with Series:

```python
s = pd.Series([1, 2, 3])
s.apply(lambda x: x**3)
```

21. Why is NumPy’s array slicing faster than Python’s list slicing
- The NumPy’s array slicing is faster than Python’s list slicing:

  - Contiguous memory storage:

      - NumPy arrays are stored in continuous blocks of memory (like C arrays).

      - Python lists store pointers to separate objects, which are scattered in memory.

      - Accessing contiguous memory is much faster due to cache efficiency.

  - Vectorized operations:

      - Slicing in NumPy returns a view of the original array, not a copy.

      - Python lists slicing creates a new list, copying all elements, which takes extra time.

  - Optimized low-level implementation:

      - NumPy is implemented in C, which is much faster for operations like slicing.

      - Python lists are handled at the Python interpreter level, which adds overhead.

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
data = {
    'A': [1, 2, 3, 4, 5],
    'B': [5, 4, 3, 2, 1],
    'C': [2, 3, 4, 5, 6]
}
df = pd.DataFrame(data)

# Calculate correlation matrix
corr_matrix = df.corr()

print("Correlation Matrix:")
print(corr_matrix)

# Visualize using a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 13. Create a 3D scatter plot using Plotly
import plotly.graph_objects as go

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 11, 12, 13, 14]
z = [5, 6, 7, 8, 9]

# Create a 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=8,
        color=z,       # Color by z value
        colorscale='Viridis',
        opacity=0.8
    )
)])

# Add title and layout details
fig.update_layout(
    title='3D Scatter Plot',
    scene=dict(
        xaxis_title='X-axis',
        yaxis_title='Y-axis',
        zaxis_title='Z-axis'
    )
)

# Show the plot
fig.show()
```
